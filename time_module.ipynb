```python
import time as t

#note : if we do not provide arguments to these functions, they will use the current time by default

curr1 = t.time() #this returns the current time in seconds since epoch (January 1, 1970)
print("Current time in seconds since epoch =", curr1)

curr2 = t.ctime() #this returns the current time in a human-readable string format
print("Current time in human-readable format =", curr2)

curr3 = t.localtime() #this returns the current local time as a struct_time object
print("Current local time as struct_time object =", curr3)
print("Year =", curr3.tm_year)
print("Month =", curr3.tm_mon)
print("Day =", curr3.tm_mday)
print("Hour =", curr3.tm_hour)
print("Minute =", curr3.tm_min)
print("Second =", curr3.tm_sec)
print("Weekday (0=Monday) =", curr3.tm_wday)

curr4 = t.mktime(curr3) #this converts a struct_time object back to seconds since epoch
print("Converted back to seconds since epoch =", curr4)


curr5 = t.gmtime() #this returns the current UTC time as a struct_time object
print("Current UTC time as struct_time object =", curr5)

curr6 = t.strftime("%Y-%m-%d %H:%M:%S", curr3) #this formats a struct_time object into a string based on the given format
print("Formatted local time string =", curr6)

curr7 = t.strptime("2023-10-01 12:30:45", "%Y-%m-%d %H:%M:%S") #this parses a string into a struct_time object based on the given format
print("Parsed struct_time object from string =", curr7)
```
