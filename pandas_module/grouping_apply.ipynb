```python
import pandas as pd
import numpy as np

data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack'],
    'Department': ['HR', 'IT', 'IT', 'Sales', 'HR', 'Sales', 'IT', 'Finance', 'Sales', 'Finance'],
    'Salary': [70000, 85000, 82000, 55000, np.nan, 58000, 91000, 95000, 62000, 88000],
    'Years_Exp': [5, 1, 7, 3, 2, 4, 10, 12, 5, 9],
    'Remote': [True, False, False, True, True, False, False, False, True, False]
}

df = pd.DataFrame(data)
display(df)
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
department = df.groupby('Department')
```

```python
department.__dict__
```

```python
department.first() # print the first entries in all the groups formed.
```

```python
for row in department:
    print(row, end='\n\n')
```

```python
grouping = df.groupby(["Department", "Remote"])
grouping.first()
```

```python
for row in grouping:
    print(row, end='\n\n')
```

```python
filtered_df = df.groupby('Department').filter(lambda x: x['Salary'].mean() >= 80000)
print(filtered_df)
```

```python
group_means = df.groupby('Department')['Salary'].mean()
result = group_means[group_means >= 50000]
print(result)
```

```python
df['Salary'] = df['Salary'].apply(lambda x : x+5000)
```
