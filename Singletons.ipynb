The __new__ method is responsible for creating the instance. By overriding it, you can check if an instance already exists before creating a new one.

```python
class Single:
    instance = None
    x = 5
    def __new__(cls):
        if cls.instance is None:
            cls.instance = super().__new__(cls)
        return cls.instance


a = Single()
a.x = 10
b = Single()

print(b.x)
```

```python
class single: 
    instance = None
    note_book = dict()
    def __new__(cls):
        if cls.instance is None:
            cls.instance = super().__new__(cls)
        return cls.instance
    
    def add_note(self, key, value):
        self.note_book[key] = value

    def get_note(self, key):
        return self.note_book.get(key, "No Note Found") 
    
note1 = single()
note1.add_note("Test User1", "999999999")

note2 = single()
note2.add_note("Test User2", "888888888")

print(note1.get_note("Test User2"))
```
