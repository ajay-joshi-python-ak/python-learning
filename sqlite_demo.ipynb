```python
import sqlite3

with sqlite3.connect('example.db') as conn: # Creates a new database file if it doesn’t exist
    cursor = conn.cursor()
    table_creation_query = """
        CREATE TABLE Employee (
            Email VARCHAR(255) NOT NULL,
            First_Name CHAR(25) NOT NULL,
            Last_Name CHAR(25)
        );
    """

    cursor.execute(table_creation_query)
```

```python
with sqlite3.connect('example.db') as conn: 
    cursor = conn.cursor()
    cursor.execute("insert into Employee values('ajayjoshi@gmail.com', 'ajay', 'joshi')")
    cursor.execute("insert into Employee values('testuser@gmail.com', 'test', 'user')")

    cursor.execute("select * from Employee")
    result = cursor.fetchall()

    for row in result: 
        print(row)
```
