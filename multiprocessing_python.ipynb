```python
import time 

def do_something(num):
    print(f"Process-{num} Doing something...")
    time.sleep(1)
    print(f"Process-{num} Done.") 

if __name__ == "__main__":
    curr1 = time.perf_counter()
    do_something(1)
    finish1 = time.perf_counter()
    print(f"Process-1 Finished in {finish1 - curr1} seconds")

    curr2 = time.perf_counter()
    do_something(2)
    finish2 = time.perf_counter()
    print(f"Process-2 Finished in {finish2 - curr2} seconds")
```

```python
import time 
import multiprocessing

def do_something(num):
    print(f"Process-{num} Doing something...")
    time.sleep(1)
    print(f"Process-{num} Done.") 

if __name__ == "__main__":
    p1 = multiprocessing.Process(target=do_something, args=(1,))
    p2 = multiprocessing.Process(target=do_something, args=(2,))

    curr = time.perf_counter()
    p1.start()
    p2.start()

    p1.join() # Used to execute rest of code after p1 is done
    p2.join() # Used to execute rest of code after p2 is done

    finish = time.perf_counter()
    print(f"Finished in {finish - curr} seconds")
```

```python
curr = time.perf_counter()

processes = []
for i in range(10):
    p = multiprocessing.Process(target=do_something, args=[i])
    processes.append(p)
    p.start()

for p in processes:
    p.join()  

finish = time.perf_counter()
print(f"Finished in {finish - curr} seconds")
```

**The executor.submit()** method in Python is used to schedule a function for asynchronous execution within an Executor (either a ThreadPoolExecutor for I/O-bound tasks or a ProcessPoolExecutor for CPU-bound tasks). It immediately returns a Future object, which represents the eventual result of the task, allowing the main program to continue running concurrently. When using concurrent.futures.submit() in Python, the task is scheduled and immediately begins executing asynchronously in the background via a ThreadPoolExecutor or ProcessPoolExecutor.

**The as_completed()** function in Python is a utility in the concurrent.futures and asyncio libraries that returns an iterator yielding futures as they complete, rather than in the order they were submitted. This allows you to process results dynamically as they become available, making programs more responsive. 

**The result()** method in Python is used to retrieve the value returned by a concurrently executed task (either a thread or a process). 

**Future.done()** It returns a boolean value (True or False) indicating whether the asynchronous execution represented by the Future object is complete.

```python
import concurrent.futures

def do_something(num):
    print(f"Process-{num} Doing something...\n")
    time.sleep(1)
    print(f"Process-{num} Done.\n")
    return num

if __name__ == "__main__":
    curr = time.perf_counter()
    with concurrent.futures.ProcessPoolExecutor() as executor:  # this will create pool of processes
        results = [executor.submit(do_something, i) for i in range(10)] # submit tasks to the pool
        
    for f in concurrent.futures.as_completed(results):
        print(f.result())
        
    finish = time.perf_counter()
    print(f"Finished in {finish - curr} seconds")
```
