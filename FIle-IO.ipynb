```python
file = open('file1.txt', 'r')
print(file.readline())
file.close()
```

```python
with open('file1.txt', 'r') as file:
    content = file.read()
    print(content, end='\n')  
    print("file mode:", file.mode)
print("File closed:", file.closed)
```

```python
with open('file1.txt', 'r') as file:
    content = file.readline()
    print(content)  
```

```python
with open('file1.txt', 'r') as file:
    for line in file:
         print(line)
```

```python
with open('image1.png', 'rb') as file:
    with open('image2.png', 'wb') as file2:
        for line in file:
            file2.write(line)  
```
