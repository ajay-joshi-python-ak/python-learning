Iterator method is a Behavioral Design Pattern that allows us to traverse the elements of the collections without taking the exposure of in-depth details of the elements. It provides a way to access the elements of complex data structure sequentially without repeating them.
In the context of the Iterator Design Pattern, you are creating a dedicated Iterator object (the class) to manage the traversal logic for a Collection object.


```python
class alphabates:
    def __init__(self, data=[]):
        self.alpha = data 
        self.index = 0
    
    def __iter__(self):
        return self 

    def __next__(self):
        if self.index<len(self.alpha):
            self.index +=1
            return self.alpha[self.index-1]
        raise Exception
    
    

obj = alphabates([chr(c) for c in range(65, 91)])

try : 
    for i in obj: 
        print(i)
except Exception:
    print("End of list")
```
