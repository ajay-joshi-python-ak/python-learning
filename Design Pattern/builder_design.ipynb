```python
class GamingPC:
    def __init__(self, cpu, ram, gpu=None, ssd=None, water_cooling=False, rgb_lighting=False):
        self.cpu = cpu
        self.ram = ram
        self.gpu = gpu
        self.ssd = ssd
        self.water_cooling = water_cooling
        self.rgb_lighting = rgb_lighting

# The Problem: 
# 1. It's hard to remember what 'True, False' refers to without looking at the class.
# 2. If you only want RGB lighting but no SSD or Water Cooling, you have to pass 
#    a bunch of 'None' or 'False' values manually.
my_pc = GamingPC("Intel i9", "32GB", "RTX 4090", None, False, True)
```

```python
class GamingPC:
    def __init__(self):
        self.cpu = None
        self.ram = None
        self.gpu = None
        self.ssd = None
        self.water_cooling = False
        self.rgb_lighting = False

    def __str__(self):
        return f"PC Config: {self.cpu}, {self.ram}, GPU: {self.gpu}, RGB: {self.rgb_lighting}"

class PCBuilder:
    def __init__(self):
        self.pc = GamingPC()

    def add_cpu(self, cpu):
        self.pc.cpu = cpu
        return self

    def add_ram(self, ram):
        self.pc.ram = ram
        return self

    def add_gpu(self, gpu):
        self.pc.gpu = gpu
        return self

    def enable_rgb(self):
        self.pc.rgb_lighting = True
        return self

    def build(self):
        return self.pc

# The Benefit:
# 1. Highly readable (Fluent Interface).
# 2. You only call the methods you need.
# 3. Order doesn't matter.
my_custom_pc = PCBuilder().add_cpu("Intel i9").add_ram("32GB").enable_rgb().build()
''' 
my_custom_pc = (
    PCBuilder()
    .add_cpu("Intel i9")
    .add_ram("32GB")
    .enable_rgb()
    .build()
)
'''

print(my_custom_pc)
```
