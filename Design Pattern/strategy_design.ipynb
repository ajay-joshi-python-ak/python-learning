The strategy method is Behavioral Design pattern that allows you to define the complete family of algorithms, encapsulates each one and putting each of them into separate classes and also allows to interchange there objects. It is implemented in Python by dynamically replacing the content of a method defined inside a class with the contents of functions defined outside of the class. It enables selecting the algorithm at run-time. This method is also called as Policy Method.

```python
class Order:
    def __init__(self, amount: float):
        self.amount = amount

    def process_payment(self, method: str, **kwargs):
        # PAIN POINT 1: Massive if/elif block that grows forever
        if method == "credit_card":
            # PAIN POINT 2: Mixing high-level logic with low-level implementation
            card_number = kwargs.get("card_number")
            print(f"Validating card {card_number}...")
            print(f"Charging ${self.amount} to Credit Card.")

        elif method == "paypal":
            email = kwargs.get("email")
            print(f"Redirecting {email} to PayPal login...")
            print(f"Charging ${self.amount} via PayPal.")

        elif method == "bitcoin":
            wallet = kwargs.get("wallet_address")
            print(f"Verifying blockchain transaction for {wallet}...")
            print(f"Charging ${self.amount} via Bitcoin.")
        
        else:
            raise ValueError("Unknown payment method")

order = Order(150.0)
order.process_payment("credit_card", card_number="1234-5678")
```

```python
from abc import ABC, abstractmethod

# 1. Strategy Interface
class PaymentStrategy(ABC):
    @abstractmethod
    def pay(self, amount: float):
        pass

# 2. Concrete Strategies
class CreditCardPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card.")

class PayPalPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paid {amount} using PayPal.")

# 3. Context
class Order:
    def __init__(self, amount: float, strategy: PaymentStrategy):
        self.amount = amount
        self.strategy = strategy

    def process(self):
        self.strategy.pay(self.amount)

# Usage
order = Order(100.0, CreditCardPayment())
order.process()  # Paid 100.0 using Credit Card.

# Switch strategy at runtime
order.strategy = PayPalPayment()
order.process() 
```
