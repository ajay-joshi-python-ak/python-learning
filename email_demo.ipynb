```python
import os 
import smtplib

Email_user = os.getenv('EMAIL_USER')
Email_pass = os.environ.get('EMAIL_PASS')

with smtplib.SMTP('smtp.gmail.com', 587) as smtp:
    smtp.ehlo()
    smtp.starttls()
    smtp.ehlo()

    smtp.login(Email_user, Email_pass)

    subject = 'Python Email Testing'
    body = 'This is a test email sent from a Python script!'

    msg = f'Subject: {subject}\n\n{body}'

    smtp.sendmail(Email_user, 'ajay.joshi@armakuni.com', msg)
```

smtp.ehlo() is a digital handshake between your Python script and the email server.

The smtp.ehlo() method in Python is part of the built-in smtplib module and is used to send an EHLO (Extended Hello) command to the SMTP server. This command identifies your client to the server and signals that your client supports the Extended SMTP (ESMTP) protocol and its features. 

**Note:** The ehlo() method is often called implicitly by methods like starttls() or sendmail() if it hasn't been called previously in the session, so explicit calls are not always strictly necessary unless you need to check specific server extensions beforehand. 

**The smtp.starttls()** method in Python's built-in smtplib library is used to put an existing insecure SMTP connection into Transport Layer Security (TLS) mode

```python
from email.message import EmailMessage
import os 
import smtplib
import imghdr

Email_user = os.environ.get('EMAIL_USER')
Email_pass = os.environ.get('EMAIL_PASS')

msg = EmailMessage()
msg['From'] = Email_user
msg['To'] = 'ajay.joshi@armakuni.com'
msg['Subject'] = 'Python Email Testing'
msg.set_content('This is a test email sent from a Python script using EmailMessage! with atteched image')
msg.add_alternative("""
                    <h1> hello</h1>
                    """, subtype='html')

with open('image1.png', 'rb') as f:
    file_data = f.read()
    file_type = imghdr.what(f.name)
    file_name = f.name 

msg.add_attachment(file_data, maintype= 'image', subtype=file_type, filename=file_name)

with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
    smtp.login(Email_user, Email_pass)
    smtp.send_message(msg)
```
